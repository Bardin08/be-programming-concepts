### Most used Pandas methods and combinations


#### Data Loading and Saving
1. **Loading Data:**
  - `pd.read_csv('file.csv')`: Load data from a CSV file.
  - `pd.read_excel('file.xlsx')`: Load data from an Excel file.
  - `pd.read_sql('SELECT * FROM table', connection)`: Load data from a SQL query.


2. **Saving Data:**
  - `df.to_csv('file.csv')`: Save DataFrame to a CSV file.
  - `df.to_excel('file.xlsx')`: Save DataFrame to an Excel file.
  - `df.to_sql('table_name', connection)`: Save DataFrame to a SQL table.


#### Data Inspection and Exploration
1. **Viewing Data:**
  - `df.head()`: View the first 5 rows.
  - `df.tail()`: View the last 5 rows.
  - `df.sample(n=5)`: View a random sample of 5 rows.


2. **Data Information:**
  - `df.info()`: Summary of the DataFrame.
  - `df.describe()`: Statistical summary of numeric columns.
  - `df.shape`: Get the dimensions of the DataFrame (rows, columns).


3. **Column and Index Information:**
  - `df.columns`: List of column names.
  - `df.index`: Index (row labels) of the DataFrame.
  - `df.dtypes`: Data types of each column.


#### Data Selection and Filtering
1. **Selecting Data:**
  - `df['column']`: Select a single column.
  - `df[['col1', 'col2']]`: Select multiple columns.
  - `df.loc[row_label, 'column']`: Select data by row label.
  - `df.iloc[row_index, col_index]`: Select data by row and column indices.


2. **Filtering Data:**
  - `df[df['column'] > value]`: Filter rows based on column condition.
  - `df[(df['col1'] > value) & (df['col2'] < value)]`: Filter rows with multiple conditions.


3. **Conditional Selection:**
  - `df.query('column > value')`: Query method for conditional selection.


#### Data Manipulation
1. **Adding/Removing Columns:**
  - `df['new_col'] = df['col1'] + df['col2']`: Add a new column.
  - `df.drop('column', axis=1)`: Drop a column.


2. **Renaming Columns:**
  - `df.rename(columns={'old_name': 'new_name'})`: Rename columns.


3. **Handling Missing Values:**
  - `df.dropna()`: Remove rows with missing values.
  - `df.fillna(value)`: Fill missing values with a specified value.


4. **Duplicating and Removing Duplicates:**
  - `df.duplicated()`: Identify duplicate rows.
  - `df.drop_duplicates()`: Remove duplicate rows.


#### Data Transformation
1. **Applying Functions:**
  - `df.apply(func)`: Apply a function along an axis of the DataFrame.
  - `df['column'].map(func)`: Map a function to a column.


2. **Group By Operations:**
  - `df.groupby('column')`: Group data by a column.
  - `df.groupby('column').agg({'col1': 'sum', 'col2': 'mean'})`: Aggregate data after grouping.


3. **Pivot Tables:**
  - `df.pivot_table(values='val', index='row', columns='col')`: Create a pivot table.


#### Data Merging and Concatenation
1. **Concatenating DataFrames:**
  - `pd.concat([df1, df2])`: Concatenate along rows.
  - `pd.concat([df1, df2], axis=1)`: Concatenate along columns.


2. **Merging DataFrames:**
  - `pd.merge(df1, df2, on='key')`: Merge DataFrames on a key column.
  - `pd.merge(df1, df2, how='left', on='key')`: Left join merge.


#### Commonly Used Method Combinations
1. **Chain Operations:**
  - **Combining Selection and Filtering:**
    ```python
    df_filtered = df[df['col1'] > value][['col2', 'col3']]
    ```
  - **Combining Grouping and Aggregation:**
    ```python
    grouped = df.groupby('col1').agg({'col2': 'sum', 'col3': 'mean'})
    ```


2. **Combining Data Cleaning Steps:**
  - **Dropping Duplicates and Handling Missing Values:**
    ```python
    df_cleaned = df.drop_duplicates().dropna()
    ```
  - **Renaming Columns and Changing Data Types:**
    ```python
    df_transformed = df.rename(columns={'old_name': 'new_name'}).astype({'col1': 'int'})
    ```

#### Example Code with Method Combinations

````python
import pandas as pd


# Load data
df = pd.read_csv('sample_data.csv')


# Inspect data
print(df.head())
print(df.info())


# Clean data
df_cleaned = df.dropna().drop_duplicates()


# Rename columns and change data type
df_cleaned = df_cleaned.rename(columns={'old_name': 'new_name'})
df_cleaned['col1'] = df_cleaned['col1'].astype(int)


# Filter data
df_filtered = df_cleaned[df_cleaned['col1'] > 50]


# Group by and aggregate
grouped = df_filtered.groupby('col2').agg({'col3': 'sum', 'col4': 'mean'})


# Save cleaned data
grouped.to_csv('cleaned_data.csv')
````

### Most Used matplotlib Methods and Combinations

#### Basic Plotting
1. **Simple Line Plot:**
   - `plt.plot(x, y)`: Create a line plot.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.show()
````

2. **Scatter Plot:**
   - `plt.scatter(x, y)`: Create a scatter plot.

````python
plt.scatter(x, y)
plt.show()
````

3. **Bar Plot:**
   - `plt.bar(x, height)`: Create a bar plot.


````python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
plt.bar(categories, values)
plt.show()
````

4. **Histogram:**
   - `plt.hist(data, bins)`: Create a histogram.

````python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.show()
````

#### Customizing Plots
1. **Titles and Labels:**
   - `plt.title('Title')`: Set the title of the plot.
   - `plt.xlabel('X-axis label')`: Set the x-axis label.
   - `plt.ylabel('Y-axis label')`: Set the y-axis label.

````python
plt.plot(x, y)
plt.title('Prime Numbers')
plt.xlabel('Index')
plt.ylabel('Prime Value')
plt.show()
````

2. **Legends:**
   - `plt.legend(['label'])`: Add a legend to the plot.

````python
plt.plot(x, y, label='Prime Numbers')
plt.legend()
plt.show()
````

3. **Colors and Styles:**
   - `plt.plot(x, y, color='r', linestyle='--')`: Customize color and line style.

````python
plt.plot(x, y, 'ro--')  # 'ro--' is a shorthand for red circles with dashed lines
plt.show()
````

4. **Grid:**
   - `plt.grid(True)`: Add grid lines to the plot.

````python
plt.plot(x, y)
plt.grid(True)
plt.show()
````

#### Subplots
1. **Creating Subplots:**
   - `plt.subplot(rows, cols, index)`: Create a subplot grid.

````python
plt.subplot(2, 1, 1)
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.bar(categories, values)

plt.show()
````

2. **Adjusting Layout:**
   - `plt.tight_layout()`: Automatically adjust subplot parameters for a neat layout.

````python
plt.subplot(2, 1, 1)
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.bar(categories, values)

plt.tight_layout()
plt.show()
````

#### Advanced Visualization Techniques
1. **Annotations:**
   - `plt.annotate('text', xy=(x, y), xytext=(x_offset, y_offset), arrowprops=dict())`: Add annotations with arrows.

````python
plt.plot(x, y)
plt.annotate('Important Point', xy=(3, 5), xytext=(3, 8), 
          arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
````

2. **Multiple Plots in One Figure:**
   - `plt.figure()`: Create a new figure.
   - `plt.plot()`: Add plots to the figure.

````python
plt.figure()
plt.plot(x, y)

plt.figure()
plt.bar(categories, values)

plt.show()
````

#### Commonly Used Method Combinations
1. **Combining Plot Types:**
   - **Line and Scatter Plot:**

````python
plt.plot(x, y, label='Line Plot')
plt.scatter(x, y, color='red', label='Scatter Plot')
plt.legend()
plt.show()
````

2. **Customizing and Adding Subplots:**
   - **Customized Subplots:**

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax1.set_title('Line Plot')

ax2.bar(categories, values)
ax2.set_title('Bar Plot')

plt.tight_layout()
plt.show()
````

#### Example Code with Method Combinations

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Create a new figure
plt.figure(figsize=(10, 5))

# Line plot with customization
plt.subplot(1, 2, 1)
plt.plot(x, y, color='blue', linestyle='--', marker='o', label='Prime Numbers')
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Prime Value')
plt.legend()
plt.grid(True)

# Bar plot with customization
plt.subplot(1, 2, 2)
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
````
